```python
def save_vocabulary(lines, filename):
    data = '\n'.join(lines)
    file = open(filename, 'w', encoding='utf-8')
    file.write(data)
    file.close()

save_vocabulary(tokens, '../static/model/vocabulary.txt')
```

```python
###vectorization
```

```python
def vectorizer(ds, vocabulary):
    vectorized_lst = []

    for sentence in ds:
        sentence_lst = np.zeros(len(vocabulary))  # Initialize a zero vector with length of vocabulary
        
        for i in range(len(vocabulary)):
            if vocabulary[i] in sentence.split():  # Check if the word is in the sentence
                sentence_lst[i] = 1  # Mark it as 1 if the word exists
        
        vectorized_lst.append(sentence_lst)  # Append the vector for the sentence

    vectorized_lst_new = np.asarray(vectorized_lst, dtype=np.float32)  # Convert the list to a NumPy array
    return vectorized_lst_new
```

```python
vectorized_x_trains=vectorizer(X_train,tokens)
```

```python
vectorized_x_test=vectorizer(X_test,tokens)
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
vectorized_x_trains_smote, y_train_smote = smote.fit_resample(vectorized_x_trains, y_train)

print(vectorized_x_trains_smote.shape, y_train_smote.shape)
```

```python
###model training
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def training_scores(y_act, y_pred):
    acc = round(accuracy_score(y_act, y_pred), 3)
    pr = round(precision_score(y_act, y_pred), 3)
    rec = round(recall_score(y_act, y_pred), 3)
    f1 = round(f1_score(y_act, y_pred), 3)
    print(f"Training Scores:\nAccuracy = {acc}\nPrecision = {pr}\nRecall = {rec}\nF1 Score = {f1}")

def validation_scores(y_act, y_pred):
    acc = round(accuracy_score(y_act, y_pred), 3)
    pr = round(precision_score(y_act, y_pred), 3)
    rec = round(recall_score(y_act, y_pred), 3)
    f1 = round(f1_score(y_act, y_pred), 3)
    print(f"Testing Scores:\nAccuracy = {acc}\nPrecision = {pr}\nRecall = {rec}\nF1-Score = {f1}")
```

```python
###Logistic Reggression
```

```python
lr= LogisticRegression()
lr.fit(vectorized_x_trains_smote,y_train_smote)
y_train_pre=lr.predict(vectorized_x_trains_smote)
y_test_pre=lr.predict(vectorized_x_test)
```

```python
##Nive bayse
```

```python
mnb= MultinomialNB()
mnb.fit(vectorized_x_trains_smote,y_train_smote)
y_train_pre=mnb.predict(vectorized_x_trains_smote)
y_test_pre=mnb.predict(vectorized_x_test)
training_scores(y_train_smote,y_train_pre)
validation_scores(y_test,y_test_pre)


```

```python
###decision tree
```

```python
dt=DecisionTreeClassifier()
dt.fit(vectorized_x_trains_smote,y_train_smote)
y_train_pre=dt.predict(vectorized_x_trains_smote)
y_test_pre=dt.predict(vectorized_x_test)
training_scores(y_train_smote,y_train_pre)
validation_scores(y_test,y_test_pre)
```

```python
###random forest
```

```python
rf=RandomForestClassifier()
rf.fit(vectorized_x_trains_smote,y_train_smote)
y_train_pre=rf.predict(vectorized_x_trains_smote)
y_test_pre=rf.predict(vectorized_x_test)
training_scores(y_train_smote,y_train_pre)
validation_scores(y_test,y_test_pre)
```
